3. Explain the following terms: The difference in np.array, np.asarray and np.asanyarray The difference between Deep copy and shallow copy.
#Ans-

**np.array, np.asarray, and np.asanyarray** are all functions used to create NumPy arrays from various input data types. While they share a common goal, they differ in their behavior and efficiency.

### np.array
 **Default behavior:** Always creates a new NumPy array, even if the input is already a NumPy array.
 **Use cases:* When we want to ensure that we have a completely independent copy of the data.

### np.asarray
 **Default behavior:** Tries to create a NumPy array without copying the input data if it's already a NumPy array.
**Use cases:** When you want to avoid unnecessary copying and ensure that the result is a NumPy array, regardless of the input type.

### np.asanyarray
 **Default behavior:** Similar to np.asarray, but it's more flexible and can handle a wider range of input types.
 **Use cases:** When you need to convert a variety of objects to NumPy arrays while avoiding unnecessary copying.

**Summary:**
 **np.array** always creates a copy.
 **np.asarray** avoids copying if possible but may not handle all input types.
 **np.asanyarray** is the most flexible and avoids copying if possible.



## Deep Copy vs. Shallow Copy

 **Shallow copy:** Creates a new object but references the same underlying data as the original object. Modifying one object will affect the other.
 **Deep copy:** Creates a new object with a completely independent copy of the data. Modifying one object will not affect the other.

**In NumPy:**
 **Array assignment:** By default, assigning one array to another creates a shallow copy.
 **np.copy(array):** Creates a deep copy of the array.
 **array.copy():** Creates a deep copy of the array.

**Approach:**
 **Shallow copy:** When you want to share data between multiple arrays and don't mind modifications to one affecting the others.
 **Deep copy:** When you need to ensure that modifications to one array don't affect others.

```python
#6. Create a 3D NumPy array of shape (3. 3, 3) containing random integers between 1 and 10. Perform the following operations. a) Find the indices of the maximum values along each depth level (third avis)b) Perform element-wise multiplication of between both array

import numpy as np

# Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10
random_array = np.random.randint(1, 11, (3, 3, 3))

# a) Find the indices of the maximum values along each depth level (third axis)
max_indices = np.unravel_index(np.argmax(random_array, axis=-1), random_array.shape[:-1])

# b) Perform element-wise multiplication of the array with itself
multiplied_array = random_array * random_array

print("Random array:\n", random_array)
print("Indices of maximum values:\n", max_indices)
print("Multiplied array:\n", multiplied_array)
```

```python
#14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

#a) Plot the Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for Temperature and right y-axis for Hurnidity).

#b) Label the x-axis as Date

#c) Set the title of the plot as Temperature and Humidity Over Time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate a time-series dataset
dates = pd.date_range(start='2023-01-01', periods=30)  # 30 consecutive days
temperature = np.random.randint(15, 35, size=30)  # Random temperatures between 15 and 35 degrees Celsius
humidity = np.random.randint(30, 90, size=30)  # Random humidity between 30% and 90%

# DataFrame with columns 'Date', 'Temperature', and 'Humidity'
df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

# 'Date' as the index for easier plotting
df.set_index('Date', inplace=True)

# a) Temperature and Humidity with different y-axes
fig, ax1 = plt.subplots()

# Temperature on the left y-axis
ax1.plot(df.index, df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# b) Create another y-axis for Humidity on the right
ax2 = ax1.twinx()
ax2.plot(df.index, df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```

```python
#15 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

#a) Plot a histogram of the data with 30 bins

#b) Overlay a line plot representing the normal distribution's pretestainty density function (PDF).

#c) Label the x-axis as Value and the y-axis as Frequency/Probability

#d) Set the title of the plot as 'Histogram with PDF Overiay

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# a) Generating 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# b) Plotting a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Generating the x values for the PDF (Probability Density Function) line
xmin, xmax = plt.xlim()  # Get current x-axis limits
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, loc=np.mean(data), scale=np.std(data))  # Calculate PDF

# Overlaing the PDF line on the histogram
plt.plot(x, pdf, 'r-', label='PDF')

# c) Labeling the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) title of the plot
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
#17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame for easier plotting with Seaborn
df = pd.DataFrame({'x': x, 'y': y})

# Assign quadrants based on x and y values
df['quadrant'] = 0
df.loc[(df['x'] >= 0) & (df['y'] >= 0), 'quadrant'] = 1  # Quadrant 1
df.loc[(df['x'] < 0) & (df['y'] >= 0), 'quadrant'] = 2  # Quadrant 2
df.loc[(df['x'] < 0) & (df['y'] < 0), 'quadrant'] = 3  # Quadrant 3
df.loc[(df['x'] >= 0) & (df['y'] < 0), 'quadrant'] = 4  # Quadrant 4

# Create the scatter plot
sns.scatterplot(x='x', y='y', hue='quadrant', data=df, palette='viridis')

# Set title and labels
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
#18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as Sine Wave Function'.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Output the plot directly into the notebook
output_notebook()

# Create data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Add grid lines (they are displayed by default, but you can modify their properties)
p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"

# Show the plot
show(p)
```

```python
#19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.


import numpy as np
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap, LinearColorMapper
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256

# Output the plot directly into the notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))

# Create a data source for Bokeh
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a new plot with title and axis labels
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value',
           tools="")

# Add bar chart with color mapping
p.vbar(x='categories', top='values', width=0.9, source=source,
       legend_field="categories", line_color='white',
       fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show the plot
show(p)
```

```python
#20  Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as Simple Line Plot.


import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo

# Generate random dataset
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100)  # Random values for y

# Create a line plot
line_plot = go.Figure()

line_plot.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update layout with title and axis labels
line_plot.update_layout(title='Simple Line Plot',
                        xaxis_title='X Axis',
                        yaxis_title='Y Axis')

# Show the plot
pyo.iplot(line_plot)
```

```python
#21    Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as Interactive Pie Chart

import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo

# Generate random data for the pie chart
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))  # Random values for each category

# Create a pie chart
pie_chart = go.Figure(data=[go.Pie(labels=categories, values=values,
                                     textinfo='label+percent',
                                     hoverinfo='label+percent')])

# Update layout with title
pie_chart.update_layout(title='Interactive Pie Chart')

# Show the plot
pyo.iplot(pie_chart)
```
